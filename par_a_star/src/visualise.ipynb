```python
import numpy as np
import matplotlib.pyplot as plt


bitmap = np.loadtxt("bitmap.txt")
mine_map = np.loadtxt("minemap.txt")

solution_map = np.loadtxt("output.txt")
bitmap = bitmap + np.transpose(solution_map)
print("Marked points in bitmap", np.sum(bitmap != 0))
print("Sum of mine map: ", np.sum(mine_map))

plt.figure(figsize=(20, 30))
plt.imshow(bitmap, cmap="inferno")
plt.colorbar()
plt.show()

plt.figure(figsize=(20, 30))
plt.imshow(mine_map, cmap="inferno")
plt.colorbar()
plt.show()

```
